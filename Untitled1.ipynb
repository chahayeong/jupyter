```python
import pandas as pd
import numpy as np
import random
import string
```

```python
def id():
    return ''.join(random.choice(string.ascii_letters) for i in range(5))
```

```python
def score():
    return np.random.randint(0,100)
```

```python
df = pd.DataFrame([[score() for i in range(1,5)] for i in range(1,11)], columns=["국어","영어","수학","사회"], index=[id() for i in range(1,11)])
df
```

```python
pd.DataFrame([[score() for i in range(1,5) ] for i in range(1,11)], 
              index=[id() for i in range(1, 11)], 
              columns=['국어','영어','수학','사회'])
```

```python
df.iloc[[0],:]
```

```python
#과학이 추가, 기존학생에게 랜덤으로 추가
df.loc[:,'과학'] = pd.Series([score() for i in range(1,11)], index=df.index)
df
```

```python
#긱 학생들의 점수의 총합을 표현하는 컬럼 추가
df.loc[:,'개인총점'] = pd.Series([], index=df.index)
                                  
```
